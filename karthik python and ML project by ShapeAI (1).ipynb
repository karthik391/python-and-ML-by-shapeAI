```python
import numpy as np
import pandas as pd
import sklearn
#now data will load the data that we will be using to train our model
#we arre using boston house pricing dataset which is available in the sklearn library itself
```

```python
from sklearn.datasets import load_boston
df=load_boston()
```

```python
df.keys() #retuen all keys of the dataset dictonary
```

```python
print(df.data)  #info about the datasets
```

```python
#we convert our dataset into the pandas dataframe,so that it is easier to analize the data

botson =pd.DataFrame(df.data,columns=df.feature_names) 
botson.head()
```

```python
#adding a new column of the target values to the dtaframe

botson['MEDV'] = df.target 
botson.head()
```

```python
#check if the datast contains any null values or not
botson.isnull()
```

```python
#checking for the null values inside it
botson.isnull().sum()
```

```python
#dividing the data into testing data and trining data
from sklearn.model_selection import train_test_split

X = botson.drop('MEDV',axis=1)
Y = botson['MEDV']

X_train, X_test, Y_train, Y_test =train_test_split(X, Y,test_size = 0.15, random_state=5)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#importing of linear regression model from sklern and train it on the training dataset

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

```

```python
#fitting model on the training datasets 

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
Y_train_predict =lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train,Y_train_predict)))

print("The model performance for training set")
print('RMSE is {}'.format(rmse))
print("\n")

Y_test_predict =lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test,Y_test_predict)))

print("The model performance for testing set")
print('RMSE is {}'.format(rmse))
```
